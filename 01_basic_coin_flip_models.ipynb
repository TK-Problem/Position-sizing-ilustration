```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from helper_functions import *
```

## Single event

### Generate event

```python
seed = 42
n_observations = 1_000
start_budget = 100
stake_pct = 0.01

# pos EV
outcomes_pos_ev = generate_binary_outcomes(0.55, n_observations=n_observations, seed=seed)
flat_model_pos_ev = flat_unit_staking(outcomes_pos_ev, stake_pct=stake_pct, start_budget=start_budget)
pct_model_pos_ev = percent_budget_staking(outcomes_pos_ev, stake_pct=stake_pct, start_budget=start_budget)

# no edge
outcomes_no_edge = generate_binary_outcomes(0.50, n_observations=n_observations, seed=seed)
flat_model_no_edge = flat_unit_staking(outcomes_no_edge, stake_pct=stake_pct, start_budget=start_budget)
pct_model_no_edge = percent_budget_staking(outcomes_no_edge, stake_pct=stake_pct, start_budget=start_budget)

# negative EV
outcomes_neg_ev = generate_binary_outcomes(0.45, n_observations=n_observations, seed=seed)
flat_model_neg_ev = flat_unit_staking(outcomes_neg_ev, stake_pct=stake_pct, start_budget=start_budget)
pct_model_neg_ev = percent_budget_staking(outcomes_neg_ev, stake_pct=stake_pct, start_budget=start_budget)
```

### Visualize results

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))
axes = axes.flatten()

# Pos EV
axes[0].plot(flat_model_pos_ev, label="Flat Staking")
axes[0].plot(pct_model_pos_ev, label="Percent Staking")
axes[0].set_title("Positive EV (p=0.55)")
axes[0].set_xlabel("Observations")
axes[0].set_ylabel("Budget")
axes[0].legend()

# No Edge
axes[1].plot(flat_model_no_edge, label="Flat Staking")
axes[1].plot(pct_model_no_edge, label="Percent Staking")
axes[1].set_title("No Edge (p=0.50)")
axes[1].set_xlabel("Observations")
axes[1].set_ylabel("Budget")
axes[1].legend()

# Negative EV
axes[2].plot(flat_model_neg_ev, label="Flat Staking")
axes[2].plot(pct_model_neg_ev, label="Percent Staking")
axes[2].set_title("Negative EV (p=0.45)")
axes[2].set_xlabel("Observations")
axes[2].set_ylabel("Budget")
axes[2].legend()

# Hide unused subplot
axes[3].axis("off")

fig.tight_layout()
```
