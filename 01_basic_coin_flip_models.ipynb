```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from helper_functions import *
```

## Single event

```python
start_budget = 100
```

```python
outcomes = generate_binary_outcomes(0.55, 1_000)
flat = flat_unit_staking(outcomes, stake_pct=0.01, start_budget=start_budget)
pct_1 = percent_budget_staking(outcomes, stake_pct=0.01, start_budget=start_budget)

plt.figure(figsize=(10, 5))
plt.plot(flat, label="Flat unit")
plt.plot(pct_1, label="1% of budget")

plt.xlabel("Bet number")
plt.ylabel("Budget")
plt.title("Staking models comparison")
plt.legend()
plt.tight_layout()
plt.show()
```
